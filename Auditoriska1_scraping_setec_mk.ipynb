```python
url = "https://setec.mk/#index.php?route=product/category&path=10002"
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
tmp = requests.get(url)
```

```python
tmp.text
```

```python
soup = BeautifulSoup(tmp.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
divs=soup.select('.name')
```

```python
produkti=[]
for div in divs:
  produkti.append(div.a.text)
```

```python
produkti
```

```python
prices_html = soup.select('.price-old-new')
```

```python
prices_html[0].text
```

```python
ceni=[]
for price in prices_html:
  ceni.append(price.text)
```

```python
len(produkti)==len(ceni)
```

```python
divs = soup.select('.right')
```

```python
divs[0].select_one('.price-old-new').text
```

```python
divs[0].select_one('.price-old-new').text
```

```python
divs[0].select_one('a').text
```

```python
lista_produkti=[]
for div in divs:
  print(div)
  if div.select_one('.price-old-new'):
    cena=div.select_one('.price-old-new').text
  else:
    cena=div.select_one('.cena_za_kesh').text
  produkt=div.select_one('a').text
  lista_produkti.append([produkt,cena])
```

```python
lista_produkti
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(lista_produkti,columns=['Produkt','Cena'])
```

```python
df
```

```python
df.to_csv('setek.csv',index=False)
```
