```python
def phones_scraping(page):
  url = "https://phones.mk/?page="+str(page)
  response = requests.get(url)
  html_doc = response.text
  soup = BeautifulSoup(html_doc, 'html.parser')
  phones = soup.select('.product-name')
  prices = soup.select('.product-price')
  telefoni=[]
  for phone in phones:
    telefoni.append(phone.text)
  for i in range(0,len(prices)):
    prices[i] = prices[i].text
  return telefoni,prices

```

```python
telefoni=[]
ceni = []
for i in range(1,11):
  tmp_tel,tmp_ceni=phones_scraping(i)
  telefoni.extend(tmp_tel)
  ceni.extend(tmp_ceni)
```

```python
len(telefoni)==len(ceni)
```
