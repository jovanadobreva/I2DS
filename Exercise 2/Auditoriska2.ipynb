```python
month = []
for date in df['date']:
  month.append(int(date.split("-")[1]))

month
```

```python
df['month']=month
```

```python
df = df.groupby('month').sum()
```

```python
df=df.reset_index()
```

```python
df['aircraft'].hist()
```

```python
import plotly.express as px
fig = px.histogram(df, x="aircraft")
fig.show()
```

```python
fig = px.bar(df, x="month", y="aircraft", color="drone", title="Aircrafts per month",barmode='stack')
fig.show()
```
