```python
import numpy as np
```

```python
X = np.zeros((150,2))
```

```python
X[:50,0] = np.random.normal(loc=0.0,scale=.3,size=50)
X[50:100,0] = np.random.normal(loc=2.0,scale=.3,size=50)
X[100:150,0] = np.random.normal(loc=-1.0,scale=.3,size=50)

X[:50,1] = np.random.normal(loc=0.0,scale=.3,size=50)
X[50:100,1] = np.random.normal(loc=2.0,scale=.3,size=50)
X[100:150,1] = np.random.normal(loc=-1.0,scale=.3,size=50)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'bo')
```

```python
from scipy.spatial.distance import cdist
```

```python
centroids = np.random.normal(loc=0.0,scale=1, size=6)
```

```python
centroids = centroids.reshape((3,2))
```

```python
cent_history = []
```

```python
cent_history=[]
cent_history.append(centroids)
for i in range(0,3):
  dist = cdist(X,centroids)
  labels = dist.argmin(axis=1)
  centroids = centroids.copy()
  centroids[0,:] = np.mean(X[labels==0,:],axis=0)
  centroids[1,:] = np.mean(X[labels==1,:],axis=0)
  centroids[2,:] = np.mean(X[labels==1,:],axis=0)
  cent_history.append(centroids)
```

```python
cent_history[0][:,1]
```

```python
for i in range(0,4):
  plt.subplot(2,2,i+1)
  dist = cdist(X,centroids)
  labels = dist.argmin(axis=1)
  plt.plot(X[labels==0,0],X[labels==0,1],'bo',label="cluster #0")
  plt.plot(X[labels==1,0],X[labels==1,1],'go',label="cluster #1")
  plt.plot(X[labels==2,0],X[labels==2,1],'co',label="cluster #2")
  plt.plot(cent_history[i][:,0],cent_history[i][:,1],'rX')
  plt.title('Step {:}'.format(i+1))
```
